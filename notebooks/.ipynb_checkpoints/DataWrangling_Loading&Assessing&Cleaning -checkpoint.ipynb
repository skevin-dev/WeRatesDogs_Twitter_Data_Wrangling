```python
df_list = []
with open('../Data/tweet-json.txt', encoding='utf-8') as file:
    lines = file.readlines()
    #print line by line
    for line in lines:
        #convert line to dict
        line_dict = json.loads(line)
        #extract tweetId, retweet count and favorite count 
        tweet_ids = line_dict['id']
        retweet_count = line_dict['retweet_count']
        favorite_count = line_dict['favorite_count']
        
        #append in created list as dictionaries 
        df_list.append({'tweet ID': tweet_ids,
                        'retweet count': retweet_count,
                        'favorite count': favorite_count})
        
df_ = pd.DataFrame(df_list, columns = ['tweet ID', 'retweet count', 'favorite count'])
```
