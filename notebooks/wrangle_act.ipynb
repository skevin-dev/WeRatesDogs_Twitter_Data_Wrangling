###### rows with multiple categories 

##### define 

There are some rows with multiple categories of dogs. To handle them we can extract their indexes and use them to drop those rows 

###### code

```python
# create empty list
multiple_category_index = []

#loop through different index and rows of twitter_archive_enhanced_clean dataset
for index, row in twitter_archive_enhanced_clean.iterrows():
    
    cat_num = 0

    if row['floofer'] == 'floofer':
        cat_num += 1
    if row['doggo'] == 'doggo':
        cat_num += 1
    if row['puppo'] == 'puppo':
        cat_num += 1
    if row['pupper'] == 'pupper':
        cat_num += 1
        #append to list index with multiple dog categories 
    if cat_num > 1:
        multiple_category_index.append(index)
```

###### test 

```python
#extract rows with the issue 
twitter_archive_enhanced_clean.loc[multiple_category_index]
```

###### code

```python
# drop the rows with multiple categories
twitter_archive_enhanced_clean.drop(multiple_category_index,axis=0,inplace=True)
```

##### Tidiness issues 

###### The four categories of dog should be combined to form a single categorical column.

###### define 

four column of dog categories are going to be converted into one column named `dog_category`

###### code

```python
#create an dog category column:
twitter_archive_enhanced_clean['dog_category'] = "No category"

#loop through rows of those categories columns and their indexes 
for i, row in twitter_archive_enhanced_clean.iterrows():
    """
    -- add a category to its corresponding dog 
    """
    if row['doggo'] == 'doggo':
        twitter_archive_enhanced_clean.at[i, 'dog_category'] = 'doggo'
    elif row['floofer'] == 'floofer':
        twitter_archive_enhanced_clean.at[i, 'dog_category'] = 'floofer'
    elif row['pupper'] == 'pupper':
        twitter_archive_enhanced_clean.at[i, 'dog_category'] = 'pupper'
    elif row['puppo'] == 'puppo':
        twitter_archive_enhanced_clean.at[i, 'dog_category'] = 'puppo'
```

###### code

```python
# we should drop the four columns and convert dog category column to categorical data type 
twitter_archive_enhanced_clean.drop(['doggo','floofer','pupper','puppo'],axis=1, inplace=True)

twitter_archive_enhanced_clean['dog_category'] = twitter_archive_enhanced_clean['dog_category'].astype('category')
```
